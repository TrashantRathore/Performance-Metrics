```python
# write your code here
data_a = pd.read_csv(r'C:\Users\trash\Downloads\Applied Assig Questions\Assignment 5(Performance Metrics)\5_a.csv')
#data_a[data_a["y"]==0.0].value_counts
data_a["y_pred"] = np.where(data_a["proba"] > 0.50, 1, 0 )
data_a.head(10)
total_negative_points = data_a[data_a["y"]==0.0].shape[0]
total_positive_points = data_a[data_a["y"]==1.0].shape[0]
tp = data_a[(data_a["y"]==1.0) & (data_a["y_pred"]==1.0)].shape[0]
fn = total_positive_points - tp
tn = data_a[(data_a["y"]==0.0) & (data_a["y_pred"]==0.0)].shape[0]
fp = total_negative_points-tn
precision = tp/(tp + fp)
recall = tp/(fn + tp)

# 1. Confusion Matrix
confusion = np.matrix([[tn,fn],[fp,tp]])
print("Confusion matrix : ","\n",confusion)

# 2. F1 Score
f1_score = (2*precision*recall)/(precision + recall)
print("F1 Score : ", f1_score)

# 3. AUC Score
data_dup = data_a
new_sorted = data_dup.sort_values(ascending=False,axis=0,by=['proba'])
gign = new_sorted.drop(columns='y_pred')

all_tpr = []
all_fpr = []
for pro_vali in gign["proba"]:
    gign["y_tilda"] = np.where(gign['proba'] >= pro_vali, 1,0 )
    total_negative_points_new = gign[gign["y"]==0.0].shape[0]
    total_positive_points_new = gign[gign["y"]==1.0].shape[0]
    tp_new = gign[(gign["y"]==1.0) & (gign["y_tilda"]==1.0)].shape[0]
    fn_new = total_positive_points_new - tp_new
    tn_new = gign[(gign["y"]==0.0) & (gign["y_tilda"]==0.0)].shape[0]
    fp_new = total_negative_points_new - tn_new
    tpr_new = tp_new/(tp_new+fn_new)
    all_tpr.append(tpr_new)
    fpr_new = fp_new/(fp_new+tn_new)
    all_fpr.append(fpr_new)
    

pauc = np.trapz(all_tpr,all_fpr)
print("AUC Score : ",pauc)

# 4. Accuracy Score
acur_score = data_a[data_a['y']==data_a['y_pred']].shape[0]
print("Accuracy Score : ",acur_score/len(data_a))

# Model performs worse than random model i.e AUC score = 0.5
# In it I have not subtracted from 1 to get AUC score > 0.5 so that we can see raw AUC score
```

<font color='red'><b>C.</b></font> Compute the best threshold (similarly to ROC curve computation) of probability which gives lowest values of metric <b>A</b> for the given data <strong>5_c.csv</strong>
<br>

you will be predicting label of a data points like this: $y^{pred}= \text{[0 if y_score < threshold  else 1]}$

$ A = 500 \times \text{number of false negative} + 100 \times \text{numebr of false positive}$

<pre>
   <b>Note 1:</b> in this data you can see number of negative points > number of positive points
   <b>Note 2:</b> use pandas or numpy to read the data from <b>5_c.csv</b>
</pre>

```python
# write your code
data_c = pd.read_csv(r'C:\Users\trash\Downloads\Applied Assig Questions\Assignment 5(Performance Metrics)\5_c.csv')
data_c



new_sorted_c = data_c.sort_values(axis=0,ascending=False,by=['prob'])

tpr_clst = []
fpr_clst = []
value_A = []
probi = []
for prob_score in new_sorted_c['prob']:
    new_sorted_c["y_tilda_new"] = np.where(new_sorted_c['prob'] >= prob_score, 1, 0 )
    total_neg_points = new_sorted_c[new_sorted_c["y"]==0.0].shape[0]
    total_pos_points= new_sorted_c[new_sorted_c["y"]==1.0].shape[0]
    tp_c = new_sorted_c[(new_sorted_c["y"] == 1.0) & (new_sorted_c["y_tilda_new"] == 1.0)].shape[0]
    fn_c = total_pos_points - tp_c
    tn_c = new_sorted_c[(new_sorted_c["y"]==0.0) & (new_sorted_c["y_tilda_new"]==0.0)].shape[0]
    fp_c = total_neg_points - tn_c
    tpr_c = tp_c/(tp_c+fn_c)
    tpr_clst.append(tpr_c)
    fpr_c = fp_c/(fp_c+tn_c)
    fpr_clst.append(fpr_c)
    A = (500 * fn_c) + (100 * fp_c)
    value_A.append(A)
    probi.append(prob_score)
    
hugg = zip(value_A,probi)


best_prob_score = min(list(hugg))[1]
print("Best Probability Score : ",best_prob_score)

data_c['y_pred'] = np.where(data_c['prob'] > best_prob_score,1,0)
data_c


# Taking best probability as threshold for y_pred
```

```python
data_d = pd.read_csv(r'C:\Users\trash\Downloads\Applied Assig Questions\Assignment 5(Performance Metrics)\5_d.csv')
data_d

# 1. Mean squared Error
#    Formula : i/n*(Summation(actual values-predicted values)^2)

data_mean = data_d
data_mean["y_pred_diff_sqr"] = (data_mean['y']-data_mean["pred"])**2
mse = np.sum(data_mean["y_pred_diff_sqr"])/len(data_mean)
print("Mean squared error : ",mse)

# 2. MAPE
#    Formula : i/n*(Summation((error(i))/actual values))

data_mean["y_pred_diff"] = abs(data_mean['y']-data_mean["pred"])
mape = np.sum(data_mean["y_pred_diff"])/np.sum(data_mean["y"])
print("Mean Absolute Percentage error : ",mape)

# 3. R^2 error
#    Fromula : 1 - SumofSquares(Residue)/SumofSquares(Total)

data_mean["total_sum_errors"] = (data_mean['y'] - np.mean(data_mean["y"]))**2
r_squared = 1 - np.sum(data_mean["y_pred_diff_sqr"])/np.sum(data_mean["total_sum_errors"])
print("R-Squared Error : ",r_squared)
data_mean
```
